#### Filling Null Values:

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer()
data = pd.DataFrame(imputer.fit_transform(df[numerical_col]),columns=numerical_col)
data.head()
```

```python
data.isnull().sum()
```

1. BALANCE: The balance left in the accounts of credit card customers.
2. PURCHASES: Amount of purchases made from the accounts of credit card customers.
3. CREDIT_LIMIT: The limit of the credit card.

These three features are enough to group credit card holders as they tell us about the buying history, bank balance, and credit limit of the credit card holders. So let’s use these features to create clusters from the dataset:

```python
clustering_data = data[["BALANCE", "PURCHASES", "CREDIT_LIMIT"]]
from sklearn.preprocessing import MinMaxScaler
for i in clustering_data.columns:
    MinMaxScaler(i)
    
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4,n_init=10)
clusters = kmeans.fit_predict(clustering_data)
data["CREDIT_CARD_SEGMENTS"] = clusters
```

```python
data.head()
```

```python
data["CREDIT_CARD_SEGMENTS"] = data["CREDIT_CARD_SEGMENTS"].map({0: "Cluster 1", 1: 
    "Cluster 2", 2: "Cluster 3", 3: "Cluster 4", 4: "Cluster 5"})
data
```

```python
plt.figure(figsize=(14,7))
sns.countplot(x='CREDIT_CARD_SEGMENTS',data=data)
plt.title('Credit Card Segments')
plt.show()
```

```python
plt.figure(figsize=(14,7))
sns.scatterplot(x='BALANCE', y='PURCHASES', data=data, hue='CREDIT_CARD_SEGMENTS', palette='rainbow')
plt.show()
```

```python
plt.figure(figsize=(14,7))
sns.scatterplot(x='BALANCE', y='CREDIT_LIMIT', data=data, hue='CREDIT_CARD_SEGMENTS', palette='rainbow')
plt.show()
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

for i in list(data["CREDIT_CARD_SEGMENTS"].unique()):
    x = data[data["CREDIT_CARD_SEGMENTS"] == i]['BALANCE']
    y = data[data["CREDIT_CARD_SEGMENTS"] == i]['PURCHASES']
    z = data[data["CREDIT_CARD_SEGMENTS"] == i]['CREDIT_LIMIT']
    
    ax.scatter(x, y, z, label=str(i))

ax.set_xlabel('BALANCE')
ax.set_ylabel('PURCHASES')
ax.set_zlabel('CREDIT_LIMIT')

ax.legend()
plt.show()
```
